```python
import os
import glob

# .txt dosyalarının bulunduğu klasör yolunu belirt
klasor_yolu = "C:/Users/edana/Desktop/newly_dataset/UAV/labels"

# Klasördeki tüm .txt dosyalarını bul
txt_dosyaları = glob.glob(os.path.join(klasor_yolu, "*.txt"))

# Her bir .txt dosyasında işlem yap
for dosya_yolu in txt_dosyaları:
    # Dosyanın içeriğini oku ve yeni bir listeye yaz
    with open(dosya_yolu, 'r') as file:
        satirlar = file.readlines()

    # Satır başındaki 2'leri 1 ile değiştir
    yeni_satirlar = []
    for satir in satirlar:
        if satir.startswith("2"):
            yeni_satirlar.append("1" + satir[1:])  # İlk karakteri değiştir
        else:
            yeni_satirlar.append(satir)

    # Değişiklikleri dosyaya geri yaz
    with open(dosya_yolu, 'w') as file:
        file.writelines(yeni_satirlar)

    print(f"{dosya_yolu} dosyasındaki işlem tamamlandı.")

print("Tüm dosyalarda işlem tamamlandı.")
```

```python
import os
import shutil
import random

# Görsel ve txt dosyalarının bulunduğu ana klasör yolları
gorsel_klasor = "D:/FINAL_DATASET/UAV/images"
txt_klasor = "D:/FINAL_DATASET/UAV/labels"

# Yeni klasör yollarını belirle
base_dir = "D:/FINAL_DATASET/UAV"
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')

# Yeni klasörleri oluştur
for directory in [train_dir, val_dir, test_dir]:
    os.makedirs(os.path.join(directory, 'images'), exist_ok=True)
    os.makedirs(os.path.join(directory, 'labels'), exist_ok=True)

# Görsel dosyalarını listele (uzantılı)
gorsel_dosyalar = [f for f in os.listdir(gorsel_klasor) if os.path.isfile(os.path.join(gorsel_klasor, f))]

# Dosyaları karıştır
random.shuffle(gorsel_dosyalar)

# Yüzdelere göre dosya sayısını hesapla
total_files = len(gorsel_dosyalar)
train_count = int(total_files * 0.7)
val_count = int(total_files * 0.2)
test_count = total_files - train_count - val_count  # Kalan dosyaları test için ayır

# Dosyaları setlere ayır
train_files = gorsel_dosyalar[:train_count]
val_files = gorsel_dosyalar[train_count:train_count + val_count]
test_files = gorsel_dosyalar[train_count + val_count:]

# Dosyaları taşıma fonksiyonu
def move_files(file_list, src_img_dir, src_txt_dir, dst_dir):
    for file_name in file_list:
        base_name, ext = os.path.splitext(file_name)
        img_src = os.path.join(src_img_dir, file_name)
        txt_src = os.path.join(src_txt_dir, f"{base_name}.txt")

        img_dst = os.path.join(dst_dir, 'images', file_name)
        txt_dst = os.path.join(dst_dir, 'labels', f"{base_name}.txt")

        if os.path.exists(img_src) and os.path.exists(txt_src):
            shutil.copy(img_src, img_dst)
            shutil.copy(txt_src, txt_dst)

# Dosyaları uygun klasörlere taşı
move_files(train_files, gorsel_klasor, txt_klasor, train_dir)
move_files(val_files, gorsel_klasor, txt_klasor, val_dir)
move_files(test_files, gorsel_klasor, txt_klasor, test_dir)

print("Dosyalar başarıyla ayrıldı!")
```
