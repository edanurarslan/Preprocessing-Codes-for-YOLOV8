```python
import os

# Görsellerin bulunduğu klasörün yolunu belirt
gorsel_klasor = "C:/Users/edana/Desktop/ADİL BİRD/drive-download-20240908T081749Z-001/images"

# Klasördeki tüm dosyaları al
for dosya in os.listdir(gorsel_klasor):
    dosya_yolu = os.path.join(gorsel_klasor, dosya)

    # Eğer dosya bir görsel dosyası ise ve uzantısı JPEG veya JPG değilse sil
    if os.path.isfile(dosya_yolu):
        uzanti = os.path.splitext(dosya)[1].lower()  # Dosya uzantısını küçük harfe çevir
        if uzanti not in ['.jpeg', '.jpg']:
            os.remove(dosya_yolu)
            print(f"{dosya} silindi.")

print("İşlem tamamlandı.")
```

```python
import os

# Görsel ve txt klasörlerinin yollarını belirt
gorsel_klasor = "C:/Users/edana/Desktop/ADİL BİRD/drive-download-20240908T081749Z-001/images"
txt_klasor = "C:/Users/edana/Desktop/ADİL BİRD/drive-download-20240908T081749Z-001/labels"

# Görsel klasöründeki dosyaları listele (uzantısız dosya adları)
gorsel_dosyalar = set(os.path.splitext(f)[0] for f in os.listdir(gorsel_klasor) if os.path.isfile(os.path.join(gorsel_klasor, f)))

# Txt klasöründeki dosyaları listele (uzantısız dosya adları)
txt_dosyalar = set(os.path.splitext(f)[0] for f in os.listdir(txt_klasor) if os.path.isfile(os.path.join(txt_klasor, f)))

# Hem görsel hem de txt klasöründe bulunan dosyalar
ortak_dosyalar = gorsel_dosyalar.intersection(txt_dosyalar)

# Görsel klasöründe olup txt klasöründe olmayan dosyaları sil
for dosya in os.listdir(gorsel_klasor):
    dosya_adi, uzanti = os.path.splitext(dosya)
    if dosya_adi not in ortak_dosyalar:
        os.remove(os.path.join(gorsel_klasor, dosya))
        print(f"{dosya} görsel klasöründen silindi.")

# Txt klasöründe olup görsel klasöründe olmayan dosyaları sil
for dosya in os.listdir(txt_klasor):
    dosya_adi, uzanti = os.path.splitext(dosya)
    if dosya_adi not in ortak_dosyalar:
        os.remove(os.path.join(txt_klasor, dosya))
        print(f"{dosya} txt klasöründen silindi.")

print("İşlem tamamlandı.")
```
