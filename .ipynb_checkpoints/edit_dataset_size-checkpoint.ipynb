```python
import os
import random

# Grsel ve txt klasrlerinin yollarn belirt
gorsel_klasor = "C:/Users/edana/Desktop/dene dataset/ZEHRA BIRD/Bird-20240908T072823Z-001/images"
txt_klasor = "C:/Users/edana/Desktop/dene dataset/ZEHRA BIRD/Bird-20240908T072823Z-001/labels"

# Grsel klasrndeki dosyalar listele (dosya adlar + uzantlar)
gorsel_dosyalar = [f for f in os.listdir(gorsel_klasor) if os.path.isfile(os.path.join(gorsel_klasor, f))]

# Txt klasrndeki dosyalar listele (dosya adlar + uzantlar)
txt_dosyalar = [f for f in os.listdir(txt_klasor) if os.path.isfile(os.path.join(txt_klasor, f))]

# Grsel dosyalarnn uzantsz isimlerini al
gorsel_dosyalar_uzantisiz = [os.path.splitext(f)[0] for f in gorsel_dosyalar]

# Txt dosyalarnn uzantsz isimlerini al
txt_dosyalar_uzantisiz = [os.path.splitext(f)[0] for f in txt_dosyalar]

# Hem grsel hem de txt klasrnde bulunan ortak dosya isimlerini bul
ortak_dosyalar = list(set(gorsel_dosyalar_uzantisiz).intersection(txt_dosyalar_uzantisiz))

# Eer ortak dosya says 500'den fazlaysa rastgele 500 dosya se
if len(ortak_dosyalar) > 500:
    ortak_dosyalar = random.sample(ortak_dosyalar, 500)

# Klasrlerden 500 dosyann dndakileri sil
for f in gorsel_dosyalar:
    dosya_adi = os.path.splitext(f)[0]
    if dosya_adi not in ortak_dosyalar:
        gorsel_dosya_yolu = os.path.join(gorsel_klasor, f)  # Tam dosya yolunu al
        if os.path.exists(gorsel_dosya_yolu):
            os.remove(gorsel_dosya_yolu)
            print(f"{gorsel_dosya_yolu} silindi.")

for f in txt_dosyalar:
    dosya_adi = os.path.splitext(f)[0]
    if dosya_adi not in ortak_dosyalar:
        txt_dosya_yolu = os.path.join(txt_klasor, f)
        if os.path.exists(txt_dosya_yolu):
            os.remove(txt_dosya_yolu)
            print(f"{txt_dosya_yolu} silindi.")

print("lem tamamland.")
```

```python
import os

# Grsel ve txt klasrlerinin yollarn belirt
gorsel_klasor = "C:/Users/edana/Desktop/dene dataset/val/images"
txt_klasor = "C:/Users/edana/Desktop/dene dataset/val/labels"

# Grsel klasrndeki dosyalar listele (uzantsz dosya adlar)
gorsel_dosyalar = set(os.path.splitext(f)[0] for f in os.listdir(gorsel_klasor) if os.path.isfile(os.path.join(gorsel_klasor, f)))

# Txt klasrndeki dosyalar listele (uzantsz dosya adlar)
txt_dosyalar = set(os.path.splitext(f)[0] for f in os.listdir(txt_klasor) if os.path.isfile(os.path.join(txt_klasor, f)))

# Ortak olmayan dosyalar kontrol et
farkli_gorseller = gorsel_dosyalar - txt_dosyalar
farkli_txtler = txt_dosyalar - gorsel_dosyalar

# Sonular yazdr
if not farkli_gorseller and not farkli_txtler:
    print("Her iki klasrde de dosya isimleri ayndr.")
else:
    if farkli_gorseller:
        print("Grsel klasrnde olup txt klasrnde olmayan dosyalar:")
        for dosya in farkli_gorseller:
            print(dosya)

    if farkli_txtler:
        print("Txt klasrnde olup grsel klasrnde olmayan dosyalar:")
        for dosya in farkli_txtler:
            print(dosya)
```

```python
import os
import random
import shutil

# Grsel ve etiket (label) dosyalarnn bulunduu klasrler
gorsel_klasor = "C:/Users/edana/Desktop/dene dataset/bird/images"
label_klasor = "C:/Users/edana/Desktop/dene dataset/bird/labels"

# Eitim, dorulama ve test dosyalarnn kaydedilecei ana dizin
hedef_klasor = "C:/Users/edana/Desktop/dene dataset/bird"

# Eitim, dorulama ve test iin alt klasr yollarn oluturun
train_dir = os.path.join(hedef_klasor, 'train')
val_dir = os.path.join(hedef_klasor, 'val')
test_dir = os.path.join(hedef_klasor, 'test')

# Alt klasrlerde de images ve labels klasrlerini olutur
for dir in [train_dir, val_dir, test_dir]:
    os.makedirs(os.path.join(dir, 'images'), exist_ok=True)
    os.makedirs(os.path.join(dir, 'labels'), exist_ok=True)

# Grsel dosyalarn (uzantlar ile birlikte) listele
gorsel_dosyalar = [f for f in os.listdir(gorsel_klasor) if os.path.isfile(os.path.join(gorsel_klasor, f))]

# Dosyalar kartr
random.shuffle(gorsel_dosyalar)

# Dosya saysna gre %70 train, %20 val, %10 test ayrma
total_files = len(gorsel_dosyalar)
train_count = int(total_files * 0.7)
val_count = int(total_files * 0.2)
test_count = total_files - train_count - val_count  # Kalan dosyalar test iin

# Dosyalar setlere ayr
train_files = gorsel_dosyalar[:train_count]
val_files = gorsel_dosyalar[train_count:train_count + val_count]
test_files = gorsel_dosyalar[train_count + val_count:]

# Dosyalar ilgili klasrlere tama fonksiyonu
def move_files(file_list, src_img_dir, src_lbl_dir, dst_dir):
    for file_name in file_list:
        base_name, ext = os.path.splitext(file_name)
        img_src = os.path.join(src_img_dir, file_name)
        lbl_src = os.path.join(src_lbl_dir, f"{base_name}.txt")

        img_dst = os.path.join(dst_dir, 'images', file_name)
        lbl_dst = os.path.join(dst_dir, 'labels', f"{base_name}.txt")

        if os.path.exists(img_src) and os.path.exists(lbl_src):
            shutil.copy(img_src, img_dst)
            shutil.copy(lbl_src, lbl_dst)

# Train, Val ve Test setlerini ilgili klasrlere ta
move_files(train_files, gorsel_klasor, label_klasor, train_dir)
move_files(val_files, gorsel_klasor, label_klasor, val_dir)
move_files(test_files, gorsel_klasor, label_klasor, test_dir)

print("Dosyalar baaryla train, val ve test klasrlerine ayrld!")
```
